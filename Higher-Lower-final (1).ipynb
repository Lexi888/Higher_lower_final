```python
account_b = random.choice(data)
score = 0
game_continue = True

while game_continue:
    def readable_account(account):
        account_name = account["name"]
        account_desc = account["description"]
        account_country = account["country"]
        return(f"{account_name},{account_desc} is from {account_country}.")
    
    account_a = account_b
    account_b = random.choice(data)
    print(f"Compare {readable_account(account_a)} vs {readable_account(account_b)}.")
    guess = input("Type your guess 'a'or 'b': ").lower()
    
    a_followers = account_a["follower_count"]
    b_followers = account_b["follower_count"]

    def check_answer(guess, a_followers, b_followers):
        if a_followers > b_followers:
            return guess == "a"
        elif a_followers == b_followers:
            return guess == "a" == "b"
        else:
            return guess == "b"
    is_correct = check_answer(guess, a_followers, b_followers)
    
    if is_correct:
        score += 1
        print(f"You win. Final score is {score}.")
    
    else:
        game_continue = False
        print(f"You lose. Final score is {score}.")
```
